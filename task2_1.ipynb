Эксперимент 3.1 провести на основе q = 1, w = 10^9, n = 1, …, 10^6 + 1 с шагом 10^4, заполнение массива a: 1 - псевдослучайное, 2 - по возрастанию, 3 - по убыванию.
(нарисовать графики функций TА(n) и ТВ(n) для всех трех случаев).

```python
import time
import copy
import random as rd
import matplotlib.pyplot as plt

from sort import k_merge_sort, d_heap_sort


q, w = 1, 1_000_000
low_n, high_n, step = 1, 1_000_000 + 1, 10_000
x_ax = [n for n in range(low_n, high_n, step)]
```

```python
""" Заполнение массива псевдослучайное """
""" создание массива случайных чисел """

arr1_r = [[]] * ((high_n - low_n) // step)
arr2_r = []
rand_arr = []

for i in range(high_n):
    rand_arr.append(rd.randint(q,w))

i = 0
for n in range(low_n, high_n, step):
    arr1_r[i] = rand_arr[:n]
    i+=1
arr2_r = copy.deepcopy(arr1_r)
```

```python
""" сортировка массивов """
heap_rand_times = []
for a in arr1_r:
    start = time.time()
    d_heap_sort(a, 3)
    end = time.time()
    heap_rand_times.append(end-start)

merge_rand_times = []
for a in arr2_r:
    start = time.time()
    k_merge_sort(a, 4)
    end = time.time()
    merge_rand_times.append(end-start)
```

```python
""" графика времени работы """
plt.plot(x_ax, merge_rand_times, 'o', markersize = 0.2)
plt.plot(x_ax, heap_rand_times, 'v', markersize = 0.2)
```

```python
""" Заполнение массива псевдослучайное """
""" создание массива числами по возрастанию """

arr1_g = [[]] * ((high_n - low_n) // step)
arr2_g = []
greate_arr = [i for i in range(q,w)]

i = 0
for n in range(low_n, high_n, step):
    arr1_g[i] = greate_arr[:n]
    i+=1
arr2_g = copy.deepcopy(arr1_g)
```

```python
""" сортировка массивов """
heap_greate_times = []
for a in arr1_g:
    start = time.time()
    d_heap_sort(a, 3)
    end = time.time()
    heap_greate_times.append(end-start)

merge_greate_times = []
for a in arr2_g:
    start = time.time()
    k_merge_sort(a, 4)
    end = time.time()
    merge_greate_times.append(end-start)
```

```python
""" график времени работы """
plt.plot(x_ax, merge_greate_times, 'o', markersize = 0.2)
plt.plot(x_ax, heap_greate_times, 'v', markersize = 0.2)
```

```python
""" Заполнение массива псевдослучайное """
""" создание массива числами по неубыванию """
arr1_l = [[]] * ((high_n - low_n) // step)
arr2_l = []
lower_arr = [i for i in range(w, q, -1)]

i = 0
for n in range(low_n, high_n, step):
    arr1_l[i] = lower_arr[:n]
    i+=1
arr2_l = copy.deepcopy(arr1_l)
```

```python
""" сортировка массивов """
heap_lower_times = []
for a in arr1_l:
    start = time.time()
    d_heap_sort(a, 3)
    end = time.time()
    heap_lower_times.append(end-start)

merge_lower_times = []
for a in arr2_l:
    start = time.time()
    k_merge_sort(a, 4)
    end = time.time()
    merge_lower_times.append(end-start)
```

```python
""" графика времени работы """
plt.plot(x_ax, merge_lower_times, 'o', markersize = 0.2)
plt.plot(x_ax, heap_lower_times, 'v', markersize = 0.2)
```

```python
""" Графики по всем 3 наполнениям массивов """
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 18))

l1, = ax1.plot(x_ax, merge_rand_times, 'o', markersize=0.3)
l2, = ax1.plot(x_ax, heap_rand_times, 'v', markersize=0.3)
ax1.legend([l1, l2], ['4 merge sort', '3 heap sort'])
ax1.set_xlabel('длина n массива')
ax1.set_ylabel('время, с')
ax1.set_title('Массив псевдослучайных чисел')

l3, = ax2.plot(x_ax, merge_greate_times, 'o', markersize=0.3)
l4, = ax2.plot(x_ax, heap_greate_times, 'v', markersize=0.3)
ax2.legend([l3, l4], ['4 merge sort', '3 heap sort'])
ax2.set_xlabel('длина n массива')
ax2.set_ylabel('время, с')
ax2.set_title('Массив чисел расположенных по возрастанию')

l5, = ax3.plot(x_ax, merge_lower_times, 'o', markersize=0.3)
l6, = ax3.plot(x_ax, heap_lower_times, 'v', markersize=0.3)
ax3.legend([l3, l4], ['4 merge sort', '3 heap sort'])
ax3.set_xlabel('длина n массива')
ax3.set_ylabel('время, с')
ax3.set_title('Массив чисел расположенных по возрастанию')
```
